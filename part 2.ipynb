# pair plot

```python
sns.pairplot(df,hue='plament')
```

# pairgrid

```python
g= sns.PairGrid(data=df,hue='plament')
g.map_diag(sns.histplot)
g.map_offdiag(sns.kdeplot)
g.map_upper(sns.boxplot)
g.map_lower(sns.barplot)
```
